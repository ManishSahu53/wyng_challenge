```python
class AutoVivification(dict):
    """Implementation of perl's autovivification feature."""
    def __getitem__(self, item):
        try:
            return dict.__getitem__(self, item)
        except KeyError:
            value = self[item] = type(self)()
            return value
```

## 1. Linear Regression

```python
def rmse(y_pred, y_true):
    sq_pred = np.power(y_pred, 2)
```

```python
def r2(y_pred, y_true):
    sq = np.sum(np.square(y_pred - y_true))
    summ = np.sum(np.square(y_true - np.mean(y_true)))
    return 1 - sq/summ
```

```python
x_train, x_test, y_train, y_test = train_test_split(store_all.values, y_pred.values, test_size=0.25)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
score = math.sqrt(mean_squared_error(y_pred, y_test))
print("RMSE of Linear Regression classifier:", round(score,4))
print("R2 of Linear Regression classifier:", np.round(r2(y_pred, y_test),4))

plt.plot(y_test, color='blue')
plt.plot(y_pred, color='red')
plt.show()
```

## 2. Decision Tree Regressor

```python
# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 0)  
  
# fit the regressor with X and Y data 
regressor.fit(x_train, y_train) 
```

```python
y_pred = regressor.predict(x_test)
score = math.sqrt(mean_squared_error(y_pred, y_test))
print("RMSE of Decision Tree Regression:", round(score,4))
print("R2 of Decision Tree Regression:", np.round(r2(y_pred, y_test),4))

plt.plot(y_test, color='blue')
plt.plot(y_pred, color='red')
plt.show()
```

## 3. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(max_depth=10, random_state=0, n_estimators=100)
forest.fit(x_train, y_train) 
```

```python
y_pred = forest.predict(x_test)
score = math.sqrt(mean_squared_error(y_pred, y_test))
print("RMSE of Random Forest classifier:", round(score,4))
print("R2 of Decision Tree classifier:", np.round(r2(y_pred, y_test),4))

plt.plot(y_test, color='blue')
plt.plot(y_pred, color='red')
plt.show()
```

## 5. AutoRegressive Model

```python
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error

# train autoregression
model = AR(y_train)
model_fit = model.fit()
print('Lag: %s' % model_fit.k_ar)
print('Coefficients: %s' % model_fit.params)
```

```python
# make predictions
predictions = model_fit.predict(start=len(y_train), end=len(y_train)+len(y_test)-1, dynamic=False)
for i in range(len(predictions)):
    print('predicted=%f, expected=%f' % (predictions[i], y_test[i]))
error = math.sqrt(mean_squared_error(y_test, predictions))
print('Test RMSE: %.3f' % error)
# plot results
plt.plot(y_test)
plt.plot(predictions, color='red')
plt.show()
```

## 6. ARIMA model

```python
from pandas import read_csv
from pandas import datetime
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA
from matplotlib import pyplot

# fit model
model = ARIMA(y_train, order=(7,1,0))
model_fit = model.fit(disp=0)
```

```python
# plot residual errors
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
residuals.plot(kind='kde')
pyplot.show()
print(residuals.describe())
```

```python
history = [x for x in y_train]
predictions = list()
for t in range(len(y_test)):
    model = ARIMA(history, order=(7,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = y_test[t]
    history.append(obs)
    print(t, len(y_test))
print('predicted=%f, expected=%f' % (yhat, obs))

error = math.sqrt(mean_squared_error(y_test, predictions))

print('Test RMSE: %.3f' % error)

# plot
plt.plot(y_test)
plt.plot(predictions, color='red')
plt.show()
```
